#### Sequência nucleotídica

```python
tfbs_dataset = read_csv('/Users/danielalemos/si/datasets/tfbs.csv', features=True, label=True)
```

```python
# kmer

kmer = KMer(k = 3, alphabet = 'DNA')
kmer_dataset = kmer.fit_transform(tfbs_dataset)
kmer_dataset.features
```

```python
# logistic regression
lg = LogisticRegression()
lg.fit(train_dataset)
```

```python
lg.score(test_dataset)
```
