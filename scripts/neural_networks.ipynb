#### Exercice 10.3

```python
ds = Dataset.from_random(n_samples = 100, n_features = 32, label = False)
y3 = np.random.randint(0,2,100)
data3 = Dataset(ds.X, y3)
data3.to_dataframe()
```

```python
layer1_3 = Dense(input_size = 32, output_size = 32)
layer2_3 = Dense(input_size = 32, output_size = 16)
layer3_3 = Dense(input_size = 16, output_size = 1)
```

```python
layer1_3_sg = SigmoidActivation ()
layer2_3_sg = SigmoidActivation ()
layer3_3_sg = SigmoidActivation ()
layers3 = [layer1_3, layer1_3_sg,layer2_3, layer2_3_sg,layer3_3, layer3_3_sg]
```

```python
nn_model_3 = NN(layers = layers3)
nn_model_3.fit(data3)
pred = nn_model_3.predict(data3)
pred
```

```python
pred = np.where(pred > 0.5, 1, 0).reshape(1,-1)
```

```python
score = accuracy(data3.y,pred)
print(score)
```

#### Exercice 12.2

```python
# datasets
breast_bin = read_csv('/Users/danielalemos/si/datasets/breast-bin.csv', features=False, label=True)
cpu = read_csv('/Users/danielalemos/si/datasets/cpu.csv', sep = ',', features = True, label = True)
```

```python
breast_bin.y[:20]
```

Binary problem

```python
# standardization
from sklearn.preprocessing import StandardScaler
breast_bin.X = StandardScaler().fit_transform(breast_bin.X)
breast_bin.X
```

```python
breast_bin_train, breast_bin_test = train_test_split(breast_bin, test_size = 0.3)
breast_bin_test.y
```

```python
print(breast_bin_train.shape())
print(breast_bin_test.shape())
```

```python
l1_bin = Dense(input_size = 9, output_size = 5)
l2_bin = Dense(input_size = 5, output_size = 1)


l1_bin_relu = ReLUActivation()
l2_bin_sig = SigmoidActivation()

layers = [l1_bin, l1_bin_relu, l2_bin, l2_bin_sig]

model_bin = NN(layers=layers, loss = cross_entropy, loss_derivative = cross_entropy_derivative, verbose = True)
```

```python
model_bin.fit(dataset = breast_bin_train)
```

```python
pred = model_bin.predict(breast_bin_test)
pred
```

```python
pred = np.where(pred > 0.5, 1, 0).reshape(1,-1)
print(pred)
```

```python
score = accuracy(breast_bin_test.y,pred)
print(score)
```

```python
l1_reg = Dense(input_size = 6, output_size = 3)
l2_reg = Dense(input_size = 3, output_size = 1)


l1_reg_relu = ReLUActivation()

layers = [l1_reg, l1_reg_relu, l2_reg]

model_reg = NN(layers = layers, loss = mse, loss_derivative = mse_derivative, verbose = True)
```

```python
from sklearn.metrics import r2_score

r2_score(cpu_test.y, pred)
```
